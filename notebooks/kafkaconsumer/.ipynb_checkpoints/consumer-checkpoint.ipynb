```python
from time import sleep       
import sys
import os

module_path = os.path.abspath('/mnt/py/modules/')
if module_path not in sys.path:
    sys.path.append(module_path)

```

```python
from KafkaConsumer import  KafkaConsumer
```

```python
#RUNNING CONSUMER FOR READING MESSAGE FROM THE KAFKA TOPIC
broker = "kafka:9092"
topic = "sellers"
group_id = "my_group"
protocol = ''
mechanism = ''
username = ''
password = ''
client_id = ''
env = 'local'

if env == 'local':
    consumer_config = {
            'bootstrap.servers': broker,
            'group.id': group_id,
            'auto.offset.reset': 'largest',
            'enable.auto.commit': 'false',
            'max.poll.interval.ms': '86400000'
        }
else:
    consumer_config = {
        'bootstrap.servers': broker,
        'group.id': group_id,
        'auto.offset.reset': 'largest',
        'enable.auto.commit': 'false',
        'max.poll.interval.ms': '86400000',
        'security.protocol': protocol,
        'sasl.mechanism': mechanism,
        'sasl.username': username,
        'sasl.password': password,
        'client.id': client_id
    }
    
  
```

```python
kafkaConsumer = KafkaConsumer(consumer_config)
consumer = kafkaConsumer.create_consumer()
```

```python
consumer.subscribe([topic])

try:
    while True:
        print("Listening")
        # read single message at a time

        msg = consumer.poll(0)

        if msg is None:
            sleep(5)
            continue
        if msg.error():
            print("Error reading message : {}".format(msg.error()))
            continue
        # You can parse message and save to data base here
        value = msg.value().decode('utf-8')
        print(value)
        consumer.commit()

except Exception as ex:
    print("Kafka Exception : {}", ex)

finally:
    print("closing consumer")
    consumer.close()


#my_consumer = ExampleConsumer()
#my_consumer.start_listener()
```
